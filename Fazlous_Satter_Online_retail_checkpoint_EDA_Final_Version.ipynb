## Capping- change the outliers values of Quantity to upper or lower limit values

```python
# Calculate the 75th and 25th percentiles of the Quantity column
upper_limit = df['Quantity'].quantile(0.75)
lower_limit = df['Quantity'].quantile(0.25)

new_df=df.copy()
new_df.loc[new_df['Quantity']>upper_limit,'Quantity']=upper_limit
new_df.loc[new_df['Quantity']<lower_limit,'Quantity']=lower_limit

# Identifying subset of data containing outliers in `Quantity`

outliers = df[(df['Quantity'] > upper_limit) | (df['Quantity'] < lower_limit)]

# Counting how many rows in the data contain outliers in `Quantity`
print("Number of rows in the data containing outliers in `Quantity`:", len(outliers))
```

## Top 20 Items that perform well in the market

```python
# Creating a Bar Chart

# Grouping by 'Items' and calculate the total 'TotalPrice' for each item\.
item_totals = new_df.groupby('Items')['TotalPrice'].sum()

# Sorting the items by total price in descending order.
top_20_items = item_totals.sort_values(ascending=False).head(20)

# Create a bar chart
plt.figure(figsize=(12, 6))  # Adjust figure size as needed
top_20_items.plot(kind='bar')
plt.title('Top 20 Items by Total Purchase Price')
plt.xlabel('Item')
plt.ylabel('Total Purchase Price')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
# Add TotalPrice as labels on top of the bars
for i, val in enumerate(top_20_items.values):
    plt.text(i, val, f'{val:,.0f}', ha='center')
plt.show()
```

## Percentage of Sale by items

```python
# Calculate the total sum of TotalPrice for the top 20 items
total_price_top20 = top_20_items.sum()

# Calculate the percentage for each of the top 20 items
percentage_by_item = (top_20_items / total_price_top20) * 100

# Create a bar chart
plt.figure(figsize=(16, 6))
bars = percentage_by_item.plot(kind='bar')
plt.title('Top 20 Items by Percentage of Total Purchase Price')
plt.xlabel('Item')
plt.ylabel('Percentage of Total Purchase Price')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Add percentage labels on top of the bars
for bar in bars.patches:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, f'{round(yval, 2)}%', ha='center', va='bottom')

plt.show()
```

## Converting 'Date'column into 'Quarter','Month','Week', and 'Day'.

```python
# Assuming 'Date' column in 'new_df' contains datetime objects
new_df['Quarter'] = new_df['Date'].dt.to_period('Q')
new_df['Month'] = new_df['Date'].dt.to_period('M')
new_df['Week'] = new_df['Date'].dt.to_period('W')
new_df['Day'] = new_df['Date'].dt.to_period('D')

# Display the updated DataFrame
print(new_df[['Date', 'Quarter', 'Month', 'Week', 'Day']].head())
```

## Monthly Sales Trend

```python
import calendar

# Extract the month from the 'Date' column
new_df['Month'] = new_df['Date'].dt.to_period('M')

# Group by month and calculate the total 'TotalPrice' for each month
monthly_performance = new_df.groupby('Month')['TotalPrice'].sum()

# Get month names from month periods
month_names = [calendar.month_name[int(str(month)[5:])] for month in monthly_performance.index]

# Create the histogram
plt.figure(figsize=(12, 6))
bars = plt.bar(month_names, monthly_performance.values)  # Use month names for x-axis
plt.xlabel('Month')
plt.ylabel('Total Purchase Price')
plt.title('Monthly Total Purchase Price Performance')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Add TotalPrice labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5, round(yval, 2), ha='center', va='bottom')

plt.show()
```

## Quarterly Trends

```python
import matplotlib.pyplot as plt

# Extract the quarter from the 'Date' column
new_df['Quarter'] = new_df['Date'].dt.to_period('Q')

# Group by quarter and calculate the total 'TotalPrice' for each quarter
quarterly_performance = new_df.groupby('Quarter')['TotalPrice'].sum()

# Create the histogram
plt.figure(figsize=(10, 6))
bars = plt.bar(quarterly_performance.index.astype(str), quarterly_performance.values)
plt.xlabel('Quarter')
plt.ylabel('Total Purchase Price')
plt.title('Quarterly Total Purchase Price Performance')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Add TotalPrice labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5, round(yval, 2), ha='center', va='bottom')

plt.show()
```

## Weekly Trends

```python
# Extract the week from the 'InvoiceDate' column
new_df['Week'] = new_df['Date'].dt.to_period('W')

# Group by week and calculate the total 'TotalPrice' for each week
weekly_performance = new_df.groupby('Week')['TotalPrice'].sum()

# Create the histogram
plt.figure(figsize=(15, 6))  # Adjust figure size for better readability
bars = plt.bar(weekly_performance.index.astype(str), weekly_performance.values)
plt.xlabel('Week')
plt.ylabel('Total Purchase Price')
plt.title('Weekly Total Purchase Price Performance')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Add TotalPrice labels on top of the bars (optional, might be cluttered for weekly data)
# for bar in bars:
#     yval = bar.get_height()
#     plt.text(bar.get_x() + bar.get_width()/2, yval + 5, round(yval, 2), ha='center', va='bottom')

plt.show()
```

## Daily Trends

```python
import matplotlib.pyplot as plt
import calendar

# Extract the day from the 'InvoiceDate' column
new_df['Day'] = new_df['Date'].dt.to_period('D')

# Group by day and calculate the total 'TotalPrice' for each day
daily_performance = new_df.groupby('Day')['TotalPrice'].sum()

# Get day names from datetime objects, ensuring all weekdays are included
all_day_names = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day_names = [calendar.day_name[day.to_timestamp().weekday()] for day in daily_performance.index]

# Calculate mean TotalPrice for each weekday, filling missing days with 0
weekday_means = daily_performance.groupby(day_names).mean().reindex(all_day_names, fill_value=0)

# Create the bar plot for mean TotalPrice by weekday
plt.figure(figsize=(10, 6))
bars = plt.bar(weekday_means.index, weekday_means.values)
plt.xlabel('Day of the Week')
plt.ylabel('Mean Total Purchase Price')
plt.title('Mean Total Purchase Price by Day of the Week')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Add mean TotalPrice labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5, round(yval, 2), ha='center', va='bottom')

plt.show()
```
